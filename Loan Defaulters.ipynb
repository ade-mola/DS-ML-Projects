```python
def parse_data(dataset):
    for column in dataset.columns:
        if dataset[column].dtype == type(object):
            le = LabelEncoder()
            dataset[column] = le.fit_transform(dataset[column].astype("str"))
    return dataset

def parse_test_data(dataset):
    for column in dataset.columns:
        if dataset[column].dtype == type(object):
            le = LabelEncoder()
            dataset[column] = le.fit_transform(dataset[column].astype("str"))
            
    return dataset

```

```python
import sklearn
from sklearn import preprocessing,metrics 
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn import linear_model,svm
from sklearn.metrics import average_precision_score
from sklearn.metrics import precision_recall_curve
```

```python
feature_cols = ['customerid', 'customerid_x', 'customerid_y', 'bank_account_type', 'bank_name_clients', 'Customer Short Address', 'systemloanid_x', 'loannumber_x', 'loanamount_x', 'totaldue_x', 'termdays', 'closeddate', 'firstduedate', 'firstrepaiddate', 'systemloanid_y', 'loannumber_y', 'loanamount_y', 'totaldue_y']
X = train[feature_cols].values
y = train.good_bad_flag.values

#features_test = test
```

```python
y.min()
```

```python
from sklearn.model_selection import GridSearchCV
def cross_validation_best_parameters(model, param_grid):
    grid = GridSearchCV(model, param_grid,cv=10, scoring='accuracy')
    X=train[feature_cols].values
    y=train.good_bad_flag.values
    grid.fit(X,y)
    mean_scores = [result.mean_validation_score for result in grid.grid_scores_]
    return mean_scores,grid.best_score_,grid.best_estimator_
logreg = linear_model.LogisticRegression(random_state=0)
c=[0.001, 0.01, 0.1, 1, 10, 100, 1000]
param_grid = dict(C=c)
mean_scores,Best_Accuracy, Best_classifier = cross_validation_best_parameters(logreg,param_grid)
print("Best accuracy is "+ str(Best_Accuracy))
print(Best_classifier)
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=20)
```

```python
clf_LR = linear_model.LogisticRegression(C=0.001)
clf_LR.fit(x_train,y_train)
#LR_Predict = clf_LR.predict_proba(x_test)[:,1]
LR_Predict_bin = clf_LR.predict(x_test)
LR_Accuracy = clf_LR.score(x_test,y_test)

print("Logistic regression accuracy is ",LR_Accuracy)

#test = LR_Predict_bin

#features_test.to_csv({LR_Accuracy}.csv,columns=["customerid","Good_Bad_Flag"],index = False)
#test
```
